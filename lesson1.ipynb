#### 1. Посмотреть документацию к API GitHub, разобраться как вывести список наименований репозиториев для конкретного пользователя, сохранить JSON-вывод в файле *.json.

```python
import requests
from pprint import pprint

url = 'https://api.github.com/users/'
user = 'PavelPoliakov' 

response = requests.get(f'{url}{user}/repos')

#полную информацию о репозиториях пользователя сохраним в переменную j_data
j_data = response.json()

pprint(j_data)
```

```python
#сделаем словарь, состоящий только из наименований репозиториев указанного пользователя и сохраним полученный словарь
#в файл *.json.
import json

repos = {
    'name': []
}

for i in j_data:
    repos['name'].append(i['name'])

with open('repos_file.json', 'w') as file:
    json.dump(repos, file)
```

```python
#выведем информацию о наименованиях репозиториев пользователя
print(f"Пользователь GitHub {user} создал следующий список репозиториев: {repos['name']}")
```
